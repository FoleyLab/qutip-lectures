```python
import json
from jc_gate import jc_gate_calc

# Load JSON data
with open('lambda_vals_1.json', 'r') as f:
    data = json.load(f)

# Store results
results = {}

# Process each entry
for key, param in data.items():
    print(f"Processing entry {key}...")
    params = param
    params['vib'] = 0.00622913
    params['S'] = 10.5615447
    params['vib_states'] = 2
    print(params)
    print(f"  lambda: {params['lambda']}, g: {params['g']}, T1: {params['T1']}, T2: {params['T2']}")
    
    # Call the calculation function
    f1, f2, conc = jc_gate_calc(params)
    
    # Store results
    results[key] = {
        'input_params': params,
        'f1': f1,
        'f2': f2,
        'concurrence': conc
    }
    
    print(f"  Results: f1={f1:.6f}, f2={f2:.6f}, conc={conc:.6f}\n")

# Save results to a new JSON file
with open('table_fidels_vib.json', 'w') as f:
    json.dump(results, f, indent=4)

print("All calculations complete! Results saved to results.json")
```
