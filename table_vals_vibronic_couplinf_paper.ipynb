```python
import numpy as np
```

```python
lams = [1e-3, 5e-3, 1e-2, 5e-2, 1e-1]
w = 0.12086
mu_eg = 1.0338263686725813

au2ev = 27.211386 # eV / a.u.

g_list = []

for lam in lams:
    g = np.sqrt(w/2) * lam * mu_eg
    g_list.append(g)

for i in range(len(g_list)):
    print(f"Lambda: {lams[i]} a.u., g: {g_list[i] * au2ev} eV")
```

```python
T1s = []
T2s = []
au2fs = 2.418884254e-2 # fs / a.u.

for g in g_list:
    T1 = np.pi / (4 * g)
    T2 = np.pi / (2 * g)

    T1s.append(T1)
    T2s.append(T2)
    
for i in range(len(lams)):
    print(f"Lambda: {lams[i]} a.u., T1: {T1s[i] * au2fs} fs, T1: {T2s[i] * au2fs} fs")
```

```python
import json

save_json = {}
for i in range(len(lams)):
    indivs = {
        'lambda': lams[i],
        'g': g_list[i],
        'T1': T1s[i],
        'T2': T2s[i]
    }
    
    save_json[i] = indivs

save_json
```

```python
with open('lambda_vals_1.json', 'w') as json_file:
        json.dump(save_json, json_file, indent=4)
```
